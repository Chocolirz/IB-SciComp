## Classical Dynamics Question 8
About the effective potential:
$$
V_{\rm eff} = \frac{J^2}{2m_er^2}-\frac{k}{r}\exp\left(-\frac{r}{a}\right).
$$
Show that circular orbits are unstable unless 
$$
\frac{r}{a} < \frac{1+\sqrt{5}}{2}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
J = 100.
m = 150.
k = 200.
a = 100.

# Define the effective energy function
def Veff(r, J, m, k, a):
    return (J**2 / (2*m*r**2)) - (k/r) * np.exp(-r/a)

# Define the range of radius r
r = np.linspace(0.135, 10, 1000)

# Plot the effective energy function
plt.figure(figsize=(8,6))
plt.title(f'Effective Energy Function')
plt.plot(r, Veff(r, J, m, k, a), label=f'Effective Energy with J={J}, m={m}, k={k}, a={a}')
plt.xlabel(r'$r$')
plt.ylabel(r'$V_{\rm eff}$')
plt.grid(True)
plt.legend()
plt.show()
```
